```python
def Extract(url):
    list = []
    for x in url:
        r = requests.get(x)
        list.append(r.text)
    return list
```

```python
data['html_text'] = Extract(url)
```

```python
def Save(text):
    title =[]
    txt = []
    for x in text:
        soup = BeautifulSoup(x, 'html.parser')
        title.append(soup.find('meta', property = 'og:title').get('content'))
        meta = (soup.find('meta', property = 'og:description'))
        if meta:
            txt.append(meta.get('content'))
        else:
            txt.append('')

    return title , txt   

    
```

```python
data['title'], data['text'] = Save(data['html_text'])
```

```python
data.head()
```

```python
stopwords = []
path = ['StopWords/StopWords_Auditor.txt','StopWords/StopWords_Currencies.txt','StopWords/StopWords_DatesandNumbers.txt','StopWords/StopWords_Generic.txt','StopWords/StopWords_GenericLong.txt','StopWords/StopWords_Geographic.txt','StopWords/StopWords_Names.txt']
for x in path:
    with open(x, 'r', errors = 'ignore') as file:
        stopwords.extend([line.strip() for line in file.readlines()])
```

```python
stopwords
```

```python
def process(text):
    pro_text = re.sub('[^a-zA-Z]',' ',text)
    pro_text = pro_text.split()
    pro_text = [word for word in pro_text if not word in stopwords]
    pro_text = ' '.join(pro_text)
    return pro_text
```

```python
data['pro_text'] = (data['text']).apply(process)
```

```python
data.head()
```

```python
positive = []
negative = []
with open('MasterDictionary//positive-words.txt','r', errors = 'ignore') as file:
    positive.extend([line.strip() for line in file.readlines()])
with open('MasterDictionary//negative-words.txt','r', errors = 'ignore') as file:
    negative.extend([line.strip() for line in file.readlines()])
```

```python
def TextAnalysis(text):
    p_score = []
    n_score = []
    for x in text:
        pos = neg =  0
        if type(x) == type(''):
            x = word_tokenize(x)
            for y in x:
                if y in positive:
                    pos += 1
                elif y in negative:
                    neg += 1
        p_score.append(pos)
        n_score.append(neg)
    return p_score, n_score
```

```python
data['POSITIVE SCORE'] , data['NEGATIVE SCORE'] = TextAnalysis(data['pro_text'])
```

```python

data['POLARITY SCORE']
data['SUBJECTIVITY']
data['AVG SENTENCE LENGTH']
data['PERCENTAGE OF COMPLEX WORDS']
data['FOG INDEX']
data['AVG NUMBER OF WORDS PER SENTENCE']
data['COMPLEX WORD COUNT']
data['WORD COUNT']
data['SYLLABLE PER WORD']
data['PERSONAL PRONOUNS']
data['AVG WORD LENGTH']
```

```python
data['POLARITY'] = ((data['POSITIVE SCORE'] - data['NEGATIVE SCORE']) / (data['POSITIVE SCORE'] + data['NEGATIVE SCORE'] + 0.000001)).round(2)
```

```python
w_count = []
s_count = []
c_count = []
for x in data['text']:
    if type(x) is type(''):
        w_count.append(len(word_tokenize(x)))
        s_count.append(x.count('.') or 1)
        count = 0
        for word in word_tokenize(x):
            count += len(word)
        c_count.append(count)
    else:
        w_count.append(0)
        s_count.append(0)
        c_count.append(0)

data['w_count'], data['s_count'], data['c_count'] = w_count, s_count, c_count
```

```python
data['SUBJECTIVITY'] = ((data['POSITIVE SCORE'] + data['NEGATIVE SCORE']) / (data['w_count'] + 0.000001)).round(2)
```

```python
data['AVG SENTENCE LENGTH'] = (data['w_count'] / data['s_count']).round(2)
```

```python
def SyllableCount(word):
    word = word.lower()
    v = 'aeiou'
    c = 0
    for x in word:
        if x in v:
            c += 1
    if word.endswith('es') or word.endswith('ed'):
        c -= 1
    return c

complex_count = []
for sent in data['pro_text']:
    count = 0
    if type(sent) is type(''):
        for word in word_tokenize(sent):
            n = SyllableCount(word)
            count += n
        complex_count.append(count)
    else:
        complex_count.append(0)

percent_complex = (((complex_count) / data['w_count']) * 100)

data['FOG INDEX'] = (0.4 * (data['AVG SENTENCE LENGTH'] + percent_complex)).round(2)

data['AVG NUMBER OF WORDS PER SENTENCE'] = (data['w_count'] / data['s_count']).round(0)

data['COMPLEX WORD COUNT'] = complex_count

data['WORD COUNT'] = w_count

data['SYLLABLE PER WORD'] = ((complex_count) / data['w_count']).round(2)

```

```python
p_noun_count = []
pronoun = ['i', 'me', 'my', 'mine', 'myself', 'you', 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'we', 'us', 'our', 'ours', 'ourselves', 'they', 'them', 'their', 'theirs', 'themselves']
for x in data['text']:
    if type(x) is type(''):
        count = 0
        for word in word_tokenize(x):
            if word.lower() in pronoun and word != 'US':
                count += 1
        p_noun_count.append(count)
    else:
        p_noun_count(0)

data['PERSONAL PRONOUNS'] = p_noun_count

data['AVG WORD LENGTH'] = (data['c_count'] / data['w_count']).round(2)
```
